```python
#j'ai scrapé le site book to scrape (notamment la catégorie des livres Fantasy) dont le lien est le  suivant: https://books.toscrape.com/


import requests
from bs4 import BeautifulSoup
```

```python
url="https://books.toscrape.com/catalogue/category/books/fantasy_19/index.html"
response=requests.get(url)
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
#extraction des livre de la première page
book_on_page_one=soup.find_all("article" , class_="product_pod")
for book in book_on_page_one:
  print(book)
```

```python
#description des différents livres de la page une
url="https://books.toscrape.com/catalogue/category/books/fantasy_19/index.html"
response=requests.get(url)
soup=BeautifulSoup(response.content,'html.parser')

#récupération de tous les url pour chaque livre
urls=[]
book_on_page_one=soup.find_all("article" , class_="product_pod")
for num_book in range(len(book_on_page_one)):
  urls.append("https://books.toscrape.com/catalogue/"+book_on_page_one[num_book].a["href"].split("/",3)[3])

description_list=[]
for book_num,url in enumerate(urls):
  response_=requests.get(url)
  soup_=BeautifulSoup(response_.content,"html.parser")
  description_list.append({f"Book {book_num+1}":soup_.find('p',class_='').string})
  # print(soup_.find('p',class_='').string)
description_list
```
