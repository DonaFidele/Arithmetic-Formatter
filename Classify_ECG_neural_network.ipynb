Il s'agit de mettre en place une approche de détection automatique des anomalies grâce aux données des électrocardiogrammes des patients dans un  nouveau centre de recherche en cardiologie.

```python
#importation des modules
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
```

```python
data= pd.read_csv('ecg.csv')
data.head()
```

```python
#données 
X=data.iloc[:,:-1]
#étiquettes
y=data.iloc[:,-1]
X
```

```python
#On vérifie que notre échantillon est bien équilibré

import plotly.graph_objects as pgo
values = y.value_counts()/y.shape[0]

fig = pgo.Figure(data=[pgo.Pie(labels=['Sain','Malade'], values=values)])
fig.update_traces(hoverinfo='label+percent', textinfo='percent', textfont_size=20,
                  marker=dict(colors=['green', 'red'], line=dict(color='white', width=0.1)))
fig.update_layout(
    title_text="Distribution des ECG sains et malades",
    title_font_color="white",
    legend_title_font_color="yellow",
    paper_bgcolor="black",
    plot_bgcolor='black',
    font_color="white",
)
fig.show()
```

```python
#division en données d'entrainement et de test
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3)
```

```python
#construction du modèle
def model_initialize():
  model=keras.Sequential()
  model.add(layers.Dense(140,activation='sigmoid',name='layer1', input_dim=2,input_shape=(140,)))
  model.add(layers.Dense(1,activation='sigmoid',name='layer2'))
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
  return model
```

```python
model = model_initialize()
model.summary()
```

```python
#entrainement du modèle
model.fit(Xtrain,ytrain,batch_size=4,epochs=50,verbose=1)
```

```python
ypred=model.predict(Xtest)
```

```python
#evaluation du modèle
f"{round(model.evaluate(Xtest,ytest)[1]*100,2)}%"
```
